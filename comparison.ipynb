# Comparison of Results at t = 0.5

This notebook compares the outputs of three different numerical and analytical approaches evaluated exactly at `t = 0.5` inside the domain `x = [0, 1]`:
1. **FEniCS** (Finite Element Method `dolfinx` using implicit formulation)
2. **PINN Neural ODE** (Physics-Informed Neural Network with `torchdiffeq`)
3. **Analytical** (Exact Fourier series Particular Integral solution)


```python
import numpy as np
import matplotlib.pyplot as plt

# Load FEM data
fem_data = np.load('fem_out.npz')
x_fem = fem_data['x']
u_fem = fem_data['u']

# Load PINNs and Analytical Data
torch_anal_data = np.load('torch_anal_out.npz')
x_ta = torch_anal_data['x']
u_pinn = torch_anal_data['u_pinn']
u_anal = torch_anal_data['u_anal']

plt.figure(figsize=(10, 6))
plt.plot(x_ta, u_anal, 'k-', linewidth=3, alpha=0.5, label='Analytical (Exact)')
plt.plot(x_fem, u_fem, 'r--', linewidth=2, label='FEM (FEniCSx)')
plt.plot(x_ta, u_pinn, 'b-.', linewidth=2, label='PINN (Neural ODE)')

plt.title('Comparison of Numerical vs Analytical Solutions (t=0.5)', fontsize=14)
plt.xlabel('Space (x)', fontsize=12)
plt.ylabel('u(x, t=0.5)', fontsize=12)
plt.legend(loc='best', fontsize=12)
plt.grid(True)
plt.show()
```
